```python
# c. Which of the following words is the longest: Supercalifragilisticexpialidocious, Honorificabilitudinitatibus, or Bababadalgharaghtakamminarronnkonn?
a = "Supercalifragilisticexpialidocious"
b = "Honorificabilitudinitatibus"
c = "Bababadalgharaghtakamminarronnkonn"

d = {a:len(a), b:len(b), c:len(c)}
k = list(d.keys())
v = list(d.values())
max_len = max(v)
lst = []
for k, v in d.items():
    if v == max_len:
        lst.append(k)
print("%s are the longest words" % (lst))
```

# Question 3



```python
import math

def triangleArea(a,b,c):
    s = (a + b + c) / 2
    area = math.sqrt(s * (s-a) * (s-b) * (s-c))
    return area

print("triangle area is : %.6f" % triangleArea(2,2,2))
```

# Question 6



```python
vowels = ['a', 'e', 'i', 'o', 'u', 'A', 'E', 'I', 'O', 'U']
def pig(str):
    if str[0] in vowels:
        return str + "way"
    else:
        return str[1:] + str[0] + "ay"
    
print(pig("happy"))
print(pig("Enter"))
```

# Question 7



```python
from collections import defaultdict

def bldcount(filename):
    dic = defaultdict(lambda: 0)
    f = open(filename, "r") 
    for line in f.readlines():
        for b in line.replace('\n','').rstrip().split(" "):
            dic[b] += 1
            
    ks = ['A', 'B', 'O', 'AB', 'OO']
    for k in ks:
        str_be = "is" if dic[k] == 1 else "are"
        cnt = "one" if dic[k] == 1 else ("no" if dic[k] == 0 else str(dic[k]))
        print("There %s %s patients of blood type %s" % (str_be, cnt, k))

bldcount('bloodtype1.txt')
```

# Question 8




```python
import pandas as pd
def curconv(cur, am):
    df = pd.read_csv('currencies.txt', names=['currency', 'amount', 'comment'], delimiter='\t')
    r = df[df['currency'] == cur]['amount'].values[0]
    return am * r

print(curconv('EUR', 100))
print(curconv('JPY', 100))
```

# Questtion 10



```python
def frequencies(str):
    str = str.lower()
    dic = defaultdict(lambda: 0)
    letters = 'abcdefghijklmnopqrstuvwxyz'
    for i in range(len(str)):
        if str[i] in letters:
            dic[str[i]] += 1
    buf = []
    for i in range(len(letters)):
        buf.append(dic[letters[i]])
    return buf

print(frequencies('The quick red fox got bored and went home.'))
print(frequencies('apple'))
```
